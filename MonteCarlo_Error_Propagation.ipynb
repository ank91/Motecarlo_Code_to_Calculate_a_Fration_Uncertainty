# Motecarlo Code to Calculate a Fration Uncertainty
Montecarlo code to compute the fraction between muon, electron and tau paticles

```python
import numpy as np

muon=[105.6583715,0.0000035]     #Mass in MeV (PDG)
elect=[0.510998928,0.000000011]  #Mass in MeV (PDG)
tau=[1776.82,0.29]               #Mass in MeV (PDG)

def density(x, y, z):
    return (x+y+z)/((x)**(1/2)+(y)**(1/2)+(z)**(1/2))**2
```

Now we generate the random masses and calculate de function

```python
m1=np.random.normal(muon[0], muon[1], 10000)
m2=np.random.normal(elect[0], elect[1], 10000)
m3=np.random.normal(tau[0], tau[1], 10000)

op=density(m1,m2,m3)
```
